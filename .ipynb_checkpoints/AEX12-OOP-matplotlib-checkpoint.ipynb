# Exercise 12 OOP matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## make a plot the same as below

```python
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
fig=plt.figure(1,figsize=(8,8))

ax3=fig.add_subplot(2,3,1)

ax3.plot(x,y1,'k')
ax3.plot(x,y2,'r')
ax3.set_xlabel('x')
ax3.set_ylabel('$\cos (x)$ and $\sin (x)$')

ax3=fig.add_subplot(2,3,3)

ax3.plot(x,y1,'k')
ax3.plot(x,y2,'r')
ax3.set_xlabel('x')
ax3.set_ylabel('$\cos (x)$ and $\sin (x)$')

ax3=fig.add_subplot(3,3,5)

ax3.plot(x,y1,'k')
ax3.plot(x,y2,'r')
ax3.set_xlabel('x')
ax3.set_ylabel('$\cos (x)$ and $\sin (x)$')
```
