# Exercise 12 OOP matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## make a plot the same as below

```python
x=np.linspace(0,2*np.pi,100)
y1=np.sin(x)
y2=np.cos(x)
fig=plt.figure(1,figsize=(8,8))

ax3=fig.add_subplot(2,1,2)

ax3.plot(x,y1,'k')
ax3.plot(x,y2,'r')
ax3.set_xlabel('x')
ax3.set_ylabel('$\cos (x)$ and $\sin (x)$')

ax1=fig.add_subplot(2,2,1)
ax1.plot(x,y1,'k')
ax1.set_xlabel('x')
ax1.set_ylabel('$\sin (x)$')

ax2=fig.add_subplot(2,2,2)
ax2.plot(x,y2,'r')
ax2.set_xlabel('x')
ax2.set_ylabel('$\cos (x)$')

fig.tight_layout()
fig.savefig('a.png')
```
